```python
# Esse arquivo contém o exericio sobre aplicação de filtros temporais

import matplotlib
import mne

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = [11., 8.]

e = '1'
trial = 3
```

```python
# Carrega o epoch sem filtros
epochs = mne.read_epochs('files/unfiltered_epo.fif')
```

```python
# Aplicando o filtro uma vez
epochs.filter(l_freq=4.0, h_freq=14.0)
epochs[e][trial].plot_psd(fmin=4.0, fmax=14.0)
print('')
```

```python
# Aplicando o filtro duas vezes
epochs.filter(l_freq=4.0, h_freq=14.0)
epochs[e][trial].plot_psd(fmin=4.0, fmax=14.0)
print('')
```

```python
# Aplicando o filtro quatro vezes
epochs.filter(l_freq=4.0, h_freq=14.0)
epochs.filter(l_freq=4.0, h_freq=14.0)
epochs[e][trial].plot_psd(fmin=4.0, fmax=14.0)
print('')
```

```python
# Aplicando o filtro oito vezes
epochs.filter(l_freq=4.0, h_freq=14.0)
epochs.filter(l_freq=4.0, h_freq=14.0)
epochs.filter(l_freq=4.0, h_freq=14.0)
epochs.filter(l_freq=4.0, h_freq=14.0)
epochs[e][trial].plot_psd(fmin=4.0, fmax=14.0)
print('')
```

```python
# Aplicando o filtro 50 vezes
for i in range(42):
    epochs.filter(l_freq=4.0, h_freq=14.0)
```

```python
epochs[e][trial].plot_psd(fmin=4.0, fmax=14.0)
print('')
```

É possível ver que aplicar o filtro várias vezes causa uma diferença no sinal, não sei o que isso significa.
